```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Survei_Makanan_Ringan.csv")
df.head()
```

```python
df = df.iloc[:, [1, 2, 4, 5, 6, 7, 8]]
df.columns = ["umur", "gender", "suka_kripik", "gizi", "suka_baru", "banyak_akses", "toko"]
df.head()
```

## Data Cleaning

```python
def clean_banyak_akses(a):
    if a[0]=='1':
        return '1';
    elif a[0] == '3':
        return '3';
    elif a[0] == '5':
        return '5';
    elif a == 'Tidak pernah':
        return '0';
    elif a == 'lebih dari 5 kali':
        return '>5'
```

```python
df['gender'] = df['gender'].apply(lambda nama:0 if (nama=='Laki - laki') else 1)
df['toko'] = df['toko'].apply(lambda toko: 0 if(toko=='Toko Online') else 1)
df['banyak_akses'] = df['banyak_akses'].apply(lambda x: clean_banyak_akses(x))
df.head()
```

```python
df['suka_kripik'] = pd.to_numeric(df['suka_kripik'])
df['gizi'] = pd.to_numeric(df['gizi'])
df['suka_baru'] = pd.to_numeric(df['suka_baru'])
```

```python
umurs = df['umur'].unique()
umurs[2], umurs[3] = umurs[3], umurs[2]
print(umurs)
```

```python
genders = df['gender'].unique()
print(genders)
```

## Data Processing

```python
data_umur_laki = []
data_umur_perem = []
nama_gender = ["Perempuan", "Laki - laki"]
iterate_rating = [1,2,3,4,5]

for umur in umurs:
    data_umur_laki.append(df[(df['gender']==0) & (df['umur']==umur)].shape[0])
    data_umur_perem.append(df[(df['gender']==1) & (df['umur']==umur)].shape[0])

print(data_umur_laki)
print(data_umur_perem)
```

```python
w = 0.4
bar1 = [0,1,2,3]
bar2 = [i+w for i in bar1]
bar_ticks = [(w/2) + i for i in bar1]

plt.figure(figsize=(16, 9))

plt.bar(bar1, data_umur_laki, w, label="Laki - laki")
plt.bar(bar2, data_umur_perem, w, label="Perempuan")

plt.title("Jumlah Responden Berdasarkan Umur dan Jenis Kelamin", fontsize=20)
plt.xticks(bar_ticks, umurs, fontsize=16)
plt.yticks(range(0, 50, 5), fontsize=16)
plt.legend(fontsize=16)

for i in range(4):
    plt.annotate(data_umur_laki[i], xy=(bar1[i], data_umur_laki[i]), 
                xytext = (bar1[i], data_umur_laki[i]+.5),
                ha='center',
                fontsize=16
                )
    
for i in range(4):
    plt.annotate(data_umur_perem[i], xy=(bar2[i], data_umur_perem[i]), 
                xytext = (bar2[i], data_umur_perem[i]+.5),
                ha='center',
                fontsize=16
                )

plt.savefig('umur_jenis.png',dpi=400)
plt.show()
```

```python
data_laki_rating = []
data_perem_rating = []
for gender in genders:
    for i in range(1,6):
        if gender == 0:
            data_laki_rating.append(df[(df['gender']==gender) & (df['suka_kripik']==i)].shape[0])
        else:
            data_perem_rating.append(df[(df['gender']==gender) & (df['suka_kripik']==i)].shape[0])   
print(data_laki_rating)
print(data_perem_rating)
```
